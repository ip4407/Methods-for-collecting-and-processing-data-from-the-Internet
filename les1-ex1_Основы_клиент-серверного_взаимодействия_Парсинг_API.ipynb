### Задание 1

Посмотреть документацию к **API GitHub**, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле `*.json`.

```python
import requests
import json
from pprint import pprint
```

```python
url = 'https://api.github.com'
user_name = 'ip4407'
```

```python
response = requests.get(url + '/users/' + user_name + '/repos')
print(url + '/users/' + user_name + '/repos')
if response.ok:
    j_data = response.json()
pprint(j_data)  # просмотр полученного содержимого
```

```python
print(f'Логин пользователя: {user_name}\n')
for i in range(0, len(j_data)):
    print(f"Имя репозитория: {j_data[i]['name']} \n Ссылка на репозиторий: {j_data[i]['svn_url']} \n")
```

```python
with open('j_data.json', 'w') as f:
    json.dump(j_data, f)
print("Данные сохранены в файл: j_data.json")
```
